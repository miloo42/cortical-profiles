```python
# images_long_df.to_csv('data/profiles_long.csv', index=False)
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

long_df = pd.read_csv('data/profiles_long.csv')
```

```python
long_df['id mod'] = long_df['brain_id'] + ' ' + long_df['modality']
```

```python
g = sns.FacetGrid(long_df, row='region', col='id mod')
g.map_dataframe(sns.lineplot, x='bin', y='profile', hue = 'section', units = 'section', estimator = None, alpha = 0.7)
g.add_legend()
g.tight_layout()
g.set_titles(col_template='{col_name}', row_template='{row_name}')

plt.savefig(f'profiles_output.png', dpi=200, bbox_inches='tight')
```
